```python
# DO NOT CHANGE
import cv2
import numpy as np
import glob
import time
## 1. Fill x_test and y_test:
x_test = []
y_test = []
fonts = ['Scheherazade New', 'Lemonada', 'Marhey','IBM Plex Sans Arabic' ]
for font in fonts:
    for filename in sorted(glob.glob(f'../../test/{font}/*.jpeg')):
        img = cv2.imread(filename, cv2.COLOR_BGR2RGB)
        x_test.append(img)
        y_test.append(fonts.index(font))

# 2. Convert them to Numpy arrays:
x_test = np.asarray(x_test)
y_test = np.asarray(y_test)

print("shape of x_data:", x_test.shape, "shape of y_data:", y_test.shape)
assert set(y_test) == {0, 1, 2, 3}
```

```python
'''
This function takes an RGB image of dimensions (1181, 1181, 3) from the test set and returns integer prediction âˆˆ {0,1,2,3}
'''
%load_ext autoreload
%autoreload 2
from preprocess import *
from SIFT import *
from classifiers_manager import *
from collections import Counter
classifiers_manager = ClassifiersManager()
classifiers_manager.load_model(Classifiers.svm)
codebook = getCodebook()
svm_model=None
with open('SVM.pkl', 'rb') as file:
    svm_model = pickle.load(file)

# Now svm_model contains the loaded SVM model
print(svm_model)
def make_prediction(img):
    # DEFINE YOUR FUNCTION HERE AND DO NOT CHANGE ANYTHING ELSE IN THE NOTEBOOK
    img=denoise(img)
    words=get_lines(img)
    DESC= prepare_data(words)
    bovw_features = compute_bovw_representation(DESC, codebook)
    bovw_features = np.array(bovw_features)
   
    predictions = svm_model.predict(bovw_features)
   
    preds_count= Counter(predictions)
    most_common_label = preds_count.most_common(1)[0][0]
    return most_common_label

# Fill your team number here
TEAM_NUM = "16"
```
